### 2. An electrical manufacturing company conducts quality checks at specified periods on the products it manufactures. Historically, the failure rate for the manufactured item is 5%. Suppose a random sample of 10 manufactured items is selected. Answer the following questions. [4 Marks]
#### 2.A. Probability that none of the items are defective? [1 Mark]

```python
#probability of failure item is 0.05
p=0.05

#random sample items
n=10

#probability of getting 0 defective items by applying binomial probability function 
#P(x) = ⁿCₓpˣ(1-p)ⁿ⁻ˣ

x=0
#prob_0_defect=(np.math.factorial(10)/(np.math.factorial(0)*np.math.factorial(10-0)))*((0.05**0)*(0.95**10))

nCx=np.math.factorial(10)/(np.math.factorial(0)*np.math.factorial(10-0))
print("nCx = {}".format(nCx))

prob_0_defects=nCx*(0.05**0)*(0.95**10)

print("Answer: probability of getting zero defects = {}".format(prob_0_defects))
bino_0_pmf=stats.binom.pmf(x,n,p)
print("Answer: probability of getting zero defects using stats= {}".format(bino_0_pmf))
```

#### 2.B. Probability that exactly one of the items is defective? [1 Mark]

```python
#probability of failure item is 0.05
p=0.05

#random sample items
n=10

#probability of getting 0 defective items by applying binomial probability function 
#P(x) = ⁿCₓpˣ(1-p)ⁿ⁻ˣ
x=1
#prob_1_defect=(np.math.factorial(10)/(np.math.factorial(1)*np.math.factorial(10-1)))*((0.05**1)*(0.95**9))

nCx=np.math.factorial(10)/(np.math.factorial(1)*np.math.factorial(10-1))
print("nCx = {}".format(nCx))

prob_1_defect=nCx*(0.05**1)*(0.95**9)

print("Answer: probability of getting 1 defect = {}".format(prob_1_defect))
bino_1_pmf=stats.binom.pmf(x,n,p)
print("Answer: probability of getting 1 defect using stats= {}".format(bino_1_pmf))
```

#### 2.C. Probability that two or fewer of the items are defective? [1 Mark]

```python
#probability of failure item is 0.05
p=0.05

#random sample items
n=10

#probability of getting 0 defective items by applying binomial probability function 
#P(x) = ⁿCₓpˣ(1-p)ⁿ⁻ˣ
x=2
#prob_2_defects=(np.math.factorial(10)/(np.math.factorial(2)*np.math.factorial(10-2)))*((0.05**2)*(0.95**8))

nCx=np.math.factorial(10)/(np.math.factorial(2)*np.math.factorial(10-2))
print("nCx = {}".format(nCx))

prob_2_defects=nCx*(0.05**2)*(0.95**8)

print("probability of getting 2 defects = {}".format(prob_2_defects))
print("probability of getting 1 defect = {}".format(prob_1_defect))
print("probability of getting 0 defects = {}".format(prob_0_defects))
prob_2_or_fewer_defects=prob_0_defects+prob_1_defect+prob_2_defects
print("Answer: probability of getting 2 or fewer defects = {}".format(prob_2_or_fewer_defects))
bino_2_cdf=stats.binom.cdf(x,n,p)
print("Answer: probability of getting 2 or fewer defects using stats= {}".format(bino_2_cdf))
```

### 3. A car salesman sells on an average 3 cars per week. [3 Marks]

#### 3.A. What is Probability that in a given week he will sell some cars? [1 Mark]

```python
rate=3
poisson_0_car=stats.poisson.pmf(0,rate)
print("probability of selling no cars = {}".format(poisson_0_car))
poisson_some_car=1-poisson_0_car
print("Answer: probability of selling some car using stats = {}".format(poisson_some_car))
```

#### 3.B. What is Probability that in a given week he will sell 2 or more but less than 5 cars? [1 Mark]

```python
rate=3
poisson_2_cars=stats.poisson.pmf(2,rate)
poisson_3_cars=stats.poisson.pmf(3,rate)
poisson_4_cars=stats.poisson.pmf(4,rate)
print("probability of selling 2 cars = {}".format(poisson_2_cars))
print("probability of selling 3 cars = {}".format(poisson_3_cars))
print("probability of selling 4 cars = {}".format(poisson_4_cars))
poisson_2more_less5_car=poisson_2_cars+poisson_3_cars+poisson_4_cars
print("Answer: probability of selling 2 or more but less than 5 cars using stats = {}".format(poisson_2more_less5_car))
```

#### 3.C. Plot the poisson distribution function for cumulative probability of cars sold per-week vs number of cars sold per week. [1 Mark]

```python
rate=3
n=np.arange(0,15)
poisson_cdf=stats.poisson.cdf(n,rate)

plt.plot(n,poisson_cdf,'o-')
plt.title('Poisson distribution for cars sold per week at rate: $\lambda$ =%i' %rate)
plt.xlabel('Number of cars sold per week')
plt.ylabel('Cumulative Probability of cars sold per week')
```

```python
#Apart from non-numeric values let us thoroughly check if there are null values

def null_check(df):
    total = df.isnull().sum().sort_values(ascending=False)   # total number of null values
    percent = (df.isnull().sum()/df.isnull().count()).sort_values(ascending=False)  # percentage of values that are null
    missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])  # putting the above two together
    return missing_data # return the dataframe
null_check(game)
```

##### Inference:
The above report shows no null values in any of the columns.

```python
#Now we can convert all types of non-numeric values to a single value - null value
#columns which have non-numeric values but expected to have numeric value
data_columns=['Score','PlayedGames','WonGames','DrawnGames','LostGames','BasketScored','BasketGiven','TournamentChampion','Runner-up']
game_final=game.drop(data_columns, axis=1).join(game[data_columns].apply(pd.to_numeric, errors='coerce'))
game_final
```

##### Inference:
The above result shows that all types of non-numeric values are converted to NaN (which is null value). game_final is the dataframe which has the converted values.

- From the distplots above for the funds raised by companies that are still operating Vs closed , the distributions seems to be similar. 
- However, to prove if there is a significant difference between the 2 groups - funds raised by operating companies Vs fund raised by closed companies, we need to do hypothesis test to compare the differences in means of the two samples. 
- Since the distributions are not normal and the sample size is > 30 , we can use 2 sample t-test or z-test in this case.

```python
#Distribution of Funds raised by all companies
sns.distplot(comp_cleaned.Funds_in_million)
plt.title('Distribution of Funds raised by all companies')
plt.show()
```

#### B. Write the null hypothesis and alternative hypothesis. [1 Mark]

Null Hypothesis - H0 : There is no significant difference between the means of the two samples i.e μ1 = μ2

Alternate Hypothesis - Ha : There is significant difference between the means of the two samples i.e μ1 != μ2

##### C. Test for significance and conclusion [1 Mark]

```python
from scipy.stats import ttest_ind
from statsmodels.stats.weightstats import ztest

sample1=comp_operating['Funds_in_million']
sample2=comp_closed['Funds_in_million']
test_statistic, p_value = ttest_ind(sample1, sample2)
print("test_statistic = {}".format(test_statistic))
print("p_value = {}".format(p_value))

alpha=0.05
print("significance level considered = {}".format(alpha))


if p_value <= alpha:
    print("Answer: p-value {0} < {1} (alpha), hence the difference is significant and we reject the Null hypothesis".format(round(p_value,3),alpha))
else:
    print("Answer: p-value {0} > {1} (alpha), hence the difference is not significant and we fail to reject the Null hypothesis".format(round(p_value,3),alpha))

#ztest_statistic, z_p_value = ztest(sample1, sample2)
#print("From z-test, test_statistic = {} and p_value ={}".format(ztest_statistic,z_p_value))
```

###### Conclusion:

##### G. Write your hypothesis comparing the proportion of companies that are operating between winners and contestants: [2 Mark]

Null hypothesis H0: The proportion of operating companies in winners category is same as that of the operating companies in contestants category.

Alternative hypothesis (Ha): The proportion of operating companies in winners category is significantly different from that of the operating companies in contestants category.

##### H. Test for significance and conclusion [1 Mark]

```python
from statsmodels.stats.proportion import proportions_ztest

test_statistic, p_value = proportions_ztest([winners_operating,contestants_operating], [winners_count,contestants_count])
print("test_statistic = {}".format(test_statistic))
print("p_value = {}".format(p_value))

alpha=0.05
print("significance level considered = {}".format(alpha))

if p_value <= alpha:
    print("Answer: p-value {0} < {1} (alpha), hence the difference is significant and we reject the Null hypothesis".format(round(p_value,3),alpha))
else:
    print("Answer: p-value {0} > {1} (alpha), hence the difference is not significant and we fail to reject the Null hypothesis".format(round(p_value,3),alpha))
```

###### Conclusion:
From the above hypothesis test, winners in the Startup Battlefield have proven to be more operational than the companies that just came out as contestants.
